```python
import pandas as pd

dummy_data = pd.read_csv("dummy.csv")
dummy_json_s = dummy_data.to_dict(orient="records")
dummy_json_s
```

```python
import requests

key = "j6xjLP5Gq9FOJib6rcJUjA"
outputs = pd.DataFrame()
for dummy in dummy_json_s:
    response = requests.post(
        "http://localhost:8080/predict",
        headers={
            "X-API-KEY": key,
            "Content-Type": "application/json",
        },
        json=dummy,
    )
    print(response.status_code)
    print(response.json())
    # concat the response to the outputs
    outputs = pd.concat([outputs, pd.DataFrame(response.json(), index=[0])])

# write the outputs to a csv
outputs.to_csv("outputs.csv", index=False)
outputs
```
